```python
import fitz  # PyMuPDF module for working with pdf documents
import re # Regular Expressions for operations on strings

def extract_and_clean_pdf(file_path):
    # Open the Portable Document Format (PDF) file
    pdf_document = fitz.open(file_path)
    print(f"Successfully opened file with {pdf_document.page_count} pages.")
    
    text_chunks = []
    
    # Iterate through each page to extract raw text
    for page_number in range(pdf_document.page_count):
        page = pdf_document.load_page(page_number)
        raw_text = page.get_text("text")
        
        # Split text by double newlines to simulate paragraph breaks
        paragraphs = raw_text.split('\n\n')
        
        for para in paragraphs:
            # Clean up messy PDF formatting: replace single newlines with spaces
            # and remove extra whitespaces, this way we make sure we don't have garbage to be processed by the BERT
            clean_para = re.sub(r'\s+', ' ', para).strip()
            
            # Filter out noise: chunks that are too short (e.g., page numbers, single words)
            if len(clean_para) > 200:
                text_chunks.append(clean_para)
                
    return text_chunks

# 1. Define path and extract
path_to_pdf = "L1.pdf" 
document_chunks = extract_and_clean_pdf(path_to_pdf)

print(f"Data preprocessed: {len(document_chunks)} valid logical chunks created.")

# Display a preview of the first chunk
if document_chunks:
    print("\n--- First Chunk Preview ---")
    print(document_chunks[0][:500] + "...")
```

```python
from sentence_transformers import SentenceTransformer # used for working with transformer neural networks
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Feature Extraction using Bidirectional Encoder Representations from Transformers (BERT)
print("Loading the 'all-MiniLM-L6-v2' BERT model...")
# it has only 6 layers, runs pretty fast on gtx 1660 TI, could be increased though
bert_model = SentenceTransformer('all-MiniLM-L6-v2')

print(f"Vectorizing {len(document_chunks)} text chunks using Graphics Processing Unit (GPU) if available...")
# The embeddings represent the mathematical features of our text
text_embeddings = bert_model.encode(document_chunks, show_progress_bar=True)

# 2. Determine Optimal Clusters using the Elbow Method (WCSS)
wcss_values = []
# Test up to 10 chapters, or the number of chunks if the document is extremely short
max_clusters_to_test = min(10, len(document_chunks))

for k in range(1, max_clusters_to_test + 1):
    # init='k-means++' ensures smarter initialization of centroids
    # n_init=10 runs the algorithm 10 times with different centroid seeds
    temp_kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    temp_kmeans.fit(text_embeddings)
    wcss_values.append(temp_kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 4))
plt.plot(range(1, max_clusters_to_test + 1), wcss_values, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Chapters')
plt.xlabel('Number of Clusters (Chapters)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid(True)
plt.show()

# 3. Apply Final K-Means Clustering
# Look at the generated plot above and change optimal_k to where the "elbow" bends.
optimal_k = 3 
print(f"\nRunning final K-Means with {optimal_k} clusters...")

final_kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)
final_kmeans.fit(text_embeddings)

chapter_assignments = final_kmeans.labels_

# 4. Display the results
print("\n--- CLUSTERING RESULTS PREVIEW ---")
for i in range(min(15, len(document_chunks))):
    print(f"[Chapter {chapter_assignments[i]}] | {document_chunks[i][:90]}...")
```
